## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

***Answer:***

Question: Which pages should a content editor review first this week — for refresh, expansion, protection, pruning, or monitoring?

Unit of analysis: one content item (one row = one pseudonymized page, content_id), trailing 90-day metrics.

Decision it improves: which page(s) get a scarce editor's attention this cycle, out of thousands of pages, when only a handful of review hours exist.

Who acts, and what they do: a content/SEO editor (or their team lead) works down a ranked queue and, for each flagged page, does one of: refresh the content, expand thin sections, protect a page that's doing well from being deprioritized, prune a page that's not worth the upkeep, or just keep monitoring it. The output isn't a probability — it's a short list with a reason attached, something a person can act on in the next sprint.

Cost of a wrong call:

False positive (flagged as "refresh first" but it wasn't worth it): a few editor-hours spent on a page that wouldn't have moved the needle — wasted labor, opportunity cost against a page that actually needed it.
False negative (a genuinely slipping, high-value page never surfaces): a client-visible page keeps losing clicks/visibility for another review cycle before anyone notices — lost traffic and, at scale, a credibility problem for the recommendation system itself.
Because both errors have real but different costs, precision and recall matter, and the ranked queue needs to surface why a page scored the way it did (a reason code), not just a bare number — so the editor can sanity-check it before spending time.
Why data or ML at all (not just a dashboard or an if-statement)? A single rule like "flag anything not updated in 90 days" ignores whether the page has any demand, whether it's visible at all, or whether it's already declining. The real signal is the interaction of several things — age, freshness, visibility, position, trend, volume — which is exactly the kind of tangled, many-signal pattern a plain if-statement handles poorly but a scored/ranked model (starting from a transparent baseline) can combine sensibly.

```python
import pandas as pd

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

n_rows, n_cols = df.shape
n_clients = df["client_id"].nunique()
pct_down = 100 * (df["trend_direction"] == "down").mean()
pct_striking = 100 * (df["position_tier"] == "striking").mean()
pct_visible = 100 * (df["avg_position"] > 0).mean()

print(f"Rows x cols: {n_rows} x {n_cols}, across {n_clients} clients")
print(f"Share of pages trending down (trend_direction == 'down'): {pct_down:.1f}%")
print(f"Share of pages in 'striking distance' position tier: {pct_striking:.1f}%")
print(f"Share of pages that are visible at all (avg_position > 0): {pct_visible:.1f}%")
```

Reading these numbers: with 30,000 pages across 32 clients, over half (~ 54%) are trending down and roughly a quarter sit in striking distance of page one — a large, overlapping pool where prioritization genuinely matters, and where a small ranking mistake (working the wrong page first) has a real opportunity cost given limited editor time. Almost all pages (~ 96%) are visible at all (avg_position > 0), so this isn't a "mostly invisible" dataset — it's a "mostly visible, often slipping" one, which is the right shape for a refresh-priority lane rather than, say, a pure discoverability/indexing lane.

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

# What this work can say:

Observed: on this 90-day window, ~54% of pages show a declining trend (trend_direction == 'down') and ~24% sit in striking distance of page one — these are measured facts about the sample, not predictions.
Directional / decision-support: a ranked queue that says "review these pages first, and here's the reason code" — a prioritization aid, not a guaranteed-correct verdict.
Model-relative: "the model ranks review-worthy pages better than a simple rule, on held-out clients" — a comparison, not an absolute truth claim.

# What this work will never say:

Not causal proof — I did not run an experiment, so I can't say refreshing a page caused a recovery, only that refreshed/high-scoring pages tend to look a certain way.
Not "predicting Google's algorithm" — I'm scoring observable content and search signals, not reverse-engineering ranking factors.
